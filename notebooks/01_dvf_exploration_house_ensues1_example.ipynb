```python
# Data directory and list all files
_project_root = Path.cwd() if (Path.cwd() / "src" / "dvf").exists() else Path.cwd().parent
raw_dir = _project_root / "data" / "raw"
all_files = sorted(raw_dir.glob("*"))
txt_files = sorted(raw_dir.glob("*.txt"))
print("All files in data/raw:", [f.name for f in all_files if not f.name.startswith(".")])
print(" .txt files (will be loaded):", [f.name for f in txt_files])
```

```python
# Load all .txt files from raw (pipe-separated), concatenate into one DataFrame
import pandas as pd

if not txt_files:
    df = pd.DataFrame()
    print("No .txt files in data/raw.")
else:
    frames = []
    for p in txt_files:
        print(f"Loading {p.name}...")
        frames.append(load_dvf_raw(p, sep="|"))
    df = pd.concat(frames, ignore_index=True)
    print(f"Total rows: {len(df):,}")
df.head()


```

```python
# Row count and display all columns (no truncation)
print(f"Total rows: {len(df):,}")
print(f"Total columns: {len(df.columns)}")
print("\nAll columns:")
display(df.columns.tolist())

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", 50)
# Preview: first 10 rows with all columns visible
display(df.head(10))
```

```python
# get min and max date (convert to datetime first; raw format is DD/MM/YYYY so string min/max was wrong)
date_col = "Date mutation"
date_series = pd.to_datetime(df[date_col], format="mixed", dayfirst=True, errors="coerce")
min_date = date_series.min()
max_date = date_series.max()
print(f"Mutation Date range: {min_date} to {max_date}")

```

```python
# Valeur fonciere is stored as "699900,00" (comma decimal) — convert before comparing
df_house1_m1["Valeur fonciere"] = pd.to_numeric(
    df_house1_m1["Valeur fonciere"].astype(str).str.replace(",", ".", regex=False),
    errors="coerce",
)
df_house1_m1 = df_house1_m1[(df_house1_m1["Valeur fonciere"] == 699900)]
```

```python
df_house1_m2 = df[(df["Date mutation"] == "06/08/2021") & (df["Section"] == "AT") 
& (df["No plan"] == 113)]
```

```python
df_house1_m2
```

```python
# Merge df_house1_m1 and df_house1_m2 (concatenate rows)
df_house1_merged = pd.concat([df_house1_m1, df_house1_m2], ignore_index=True)
print(f"Merged dataframe: {len(df_house1_merged)} rows")
print(f"  - df_house1_m1: {len(df_house1_m1)} rows")
print(f"  - df_house1_m2: {len(df_house1_m2)} rows")
df_house1_merged
```

```python
# keep only from df_house1_merged the rows with Type local = MAISON
df_house1_merged = df_house1_merged[df_house1_merged["Type local"] == "Maison"]
df_house1_merged

```

```python
display.max_columns=None
display.max_colwidth = None
```

```python
# Create summary dataframe with house details and mutations as JSON string
import json

# 1. Choose which columns describe the house (same for all mutations)
house_cols = [
    "Code postal",
    "Commune",
    "Code departement",
    "Code commune",
    "Section",
    "No plan",
    "Type local",
    "Surface reelle bati",
    "Surface terrain",
    "Nombre pieces principales"
]

# 2. Take house info from the first row (they all refer to the same house)
house_info = df_house1_merged.iloc[0][house_cols].to_dict()

# 3. Build mutations as list of key-value pairs: Date mutation -> Valeur fonciere
# Each mutation becomes {"Date mutation": "Valeur fonciere"}
mutations_list = [
    {row["Date mutation"]: str(row["Valeur fonciere"])}
    for _, row in df_house1_merged.iterrows()
]

# 4. Convert to JSON string
mutations_json = json.dumps(mutations_list, ensure_ascii=False)

# 5. Create a single-row dataframe with mutations as JSON string
df_house_summary = pd.DataFrame([{
    **house_info,
    "mutations": mutations_json,
}])

print(f"Mutations JSON string: {mutations_json}")
print(f"\nType: {type(mutations_json)}")
df_house_summary
```

```python
# print df_house_summary "mutations" field in entirety
mutations_str = df_house_summary["mutations"].iloc[0]
mutations_parsed = json.loads(mutations_str)
print(json.dumps(mutations_parsed, indent=2, ensure_ascii=False))
```
