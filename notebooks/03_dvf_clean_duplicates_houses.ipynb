```python
# clean duplicates in the dataset data/processed/df_grouped_ensues_2020_2025_houses.csv
# keep only the first occurrence of each duplicate for the field mutations.
# save the cleaned dataset to data/processed/df_grouped_ensues_2020_2025_houses_cleaned.csv

import pandas as pd
import ast

# Load the dataset
input_path = "../data/processed/df_grouped_2020_2025_france.csv"
output_path = "../data/processed/df_grouped_2020_2025_france_cleaned.csv"

df = pd.read_csv(input_path)

def clean_mutations(mutations_str):
    """
    Remove duplicates from the list of mutations, keeping only the first occurrence of each.
    The input is assumed to be a string representation of a list of dicts: '[{"date": "price"}, ...]'
    """
    try:
        mutations = ast.literal_eval(mutations_str)
        seen = set()
        cleaned = []
        for mutation in mutations:
            # Each mutation is a dict with one date: price
            # We'll use tuple (date, price) as the hashable key
            if isinstance(mutation, dict):
                for date, price in mutation.items():
                    key = (date, price)
                    if key not in seen:
                        cleaned.append({date: price})
                        seen.add(key)
        return str(cleaned)
    except Exception as e:
        # If parsing fails, return the original string
        return mutations_str

df['mutations'] = df['mutations'].apply(clean_mutations)

df.to_csv(output_path, index=False)
print(f"Saved cleaned data to {output_path}")



```

```python
# Filter to see Ensues-la-Redonne, Section AT, No plan 113
# 13820 = Code postal (not Code commune; Code commune for Ensues = 33)
mask_cp = pd.to_numeric(df["Code postal"], errors="coerce") == 13820
mask_sec = df["Section"].astype(str).str.strip() == "AT"
mask_plan = pd.to_numeric(df["No plan"], errors="coerce") == 113
df_ensues_at_113 = df[mask_cp & mask_sec & mask_plan]
print(f"Rows for Ensues (13820), Section AT, No plan 113: {len(df_ensues_at_113)}")
pd.set_option("display.max_colwidth", None)
df_ensues_at_113
```
